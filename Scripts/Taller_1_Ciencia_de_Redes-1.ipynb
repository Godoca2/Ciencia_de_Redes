```python
# Importación de las librerías necesarias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```


### i) Desarrollo de la red

En esta etapa, se definió la estructura base del grafo mediante la identificación de los siguientes elementos clave:

- **Nodos:** Representan las entidades individuales de la red.
- **Enlaces:** Se establecieron las conexiones entre nodos, representando las relaciones existentes.
- **Pesos de los enlaces:** A cada conexión se le asignó un valor numérico, indicando la intensidad o relevancia de la relación.

Este paso sentó las bases para la construcción del grafo y permitió formalizar la representación matemática de la red.


```python
# Grafo
G = nx.Graph()
```

```python
# Nodo
nodos_list = [1,2,3,4,5]
G.add_nodes_from(nodos_list)
G.nodes(data=True)
```

```python
# Enlaces
edge_list = [(1,2,0.5), # u,v, peso
            (1,3,0.2),
            (1,5,0.3),
            (2,3,0.4),
            (2,4,0.7),
            (2,5,0.4),
            (3,4,0.2),
            (3,5,0.2),
            (4,5,0.1)]
G.add_weighted_edges_from(edge_list)
```

```python
# Nodos y enlaces (mejora en la presentación)

print("Nodos:")
for node in G.nodes:
    print(f"- Nodo: {node}")

print("\nEnlaces con pesos:")
for u, v, data in G.edges(data=True):
    print(f"- Enlace: {u} <--> {v}, Peso: {data['weight']}")
```

### ii) Matriz de adyacencia

La matriz de adyacencia es una representación bidimensional de un grafo, donde:

- Las **filas** y **columnas** representan los nodos de la red.
- Los **valores en las celdas** indican la existencia de un enlace entre los nodos:

  - Un valor diferente de cero representa un enlace, mientras que un valor de cero indica la ausencia de conexión.
  - En grafos ponderados, el valor en la celda corresponde al peso del enlace entre los nodos.

En esta etapa, se generó la matriz de adyacencia correspondiente al grafo, permitiendo una representación matemática que facilita su análisis y manipulación. Esta matriz constituye la base para entender la conectividad y evaluar las propiedades estructurales de la red.



```python
# Matriz de adyacencia mejorada
import seaborn as sns
import matplotlib.pyplot as plt

# Convertir la matriz de adyacencia a DataFrame
adj_matrix = nx.adjacency_matrix(G).todense()
adj_df = pd.DataFrame(adj_matrix, index=G.nodes, columns=G.nodes)

# Mostrar matriz de adyacencia como tabla
print("Matriz de Adyacencia:")
print(adj_df)

# Visualización de la matriz de adyacencia
plt.figure(figsize=(8, 6))
sns.heatmap(adj_df, annot=True, fmt=".1f", cmap="Blues", cbar=True)
plt.title("Matriz de Adyacencia del Grafo")
plt.xlabel("Nodos")
plt.ylabel("Nodos")
plt.show()

# Exportar matriz de adyacencia como archivo CSV (opcional)
# adj_df.to_csv("matriz_adyacencia.csv", index=True)
```

### Interpretación de la matriz de adyacencia

El mapa de calor de la matriz de adyacencia nos permite analizar las relaciones entre los nodos de la red de manera visual e intuitiva. A continuación, se detallan los aspectos clave de la interpretación:

1. **Ejes del mapa**:
   - Las filas y columnas representan los nodos del grafo.
   - La intersección entre una fila (nodo origen) y una columna (nodo destino) muestra el peso del enlace entre esos nodos.

2. **Valores representados**:
   - Los valores dentro de la matriz indican la existencia y la intensidad de las conexiones entre los nodos.
   - Un valor de **0.0** (representado en blanco) indica que no existe un enlace entre los nodos.
   - Valores mayores, como **0.7** (representado en azul oscuro), indican relaciones más fuertes entre los nodos.

3. **Simetría de la matriz**:
   - La matriz es simétrica respecto a la diagonal principal, ya que el grafo es no dirigido. Esto implica que la relación entre un nodo A y un nodo B es igual a la relación entre el nodo B y el nodo A.

4. **Escala de colores**:
   - La intensidad del color aumenta con el peso de las conexiones. 
   - Los colores claros (cercanos a blanco) representan valores bajos (0.0), mientras que los colores oscuros (cercanos a azul profundo) representan valores altos (hasta 0.7).

5. **Ejemplo de relaciones observadas**:
   - La conexión entre los nodos **2 y 4** tiene un peso de **0.7**, siendo la relación más fuerte en la red.
   - La conexión entre los nodos **1 y 3** tiene un peso de **0.2**, mostrando una relación más débil en comparación con otras.
   - Los nodos que no tienen conexiones entre ellos, como **1 y 4**, se representan con celdas de valor **0.0**.

6. **Usos del mapa**:
   - Identificación de patrones en la red, como nodos muy conectados o nodos aislados.
   - Visualización rápida de las relaciones más importantes en la red para facilitar su análisis.

Esta representación permite comprender de manera clara las conexiones de la red y destaca los pesos de las relaciones, proporcionando una base sólida para análisis posteriores.


### iii) Visualización de la red

En esta etapa se realizaron dos visualizaciones con el objetivo de analizar y comparar la red generada con la red objetivo proporcionada. Los pasos realizados son los siguientes:

1. **Visualización inicial de la red generada**:
   - Se utilizó una representación gráfica para mostrar la estructura inicial de la red generada.
   - Los nodos y enlaces se visualizaron con sus respectivos pesos, destacando las conexiones establecidas en la matriz de adyacencia.

2. **Ajustes para replicar la red objetivo**:
   - Se ajustaron las posiciones de los nodos para que la red generada coincidiera con la disposición visual de la red objetivo.
   - Se personalizó el diseño para incluir:
     - Etiquetas de los nodos y pesos de los enlaces.
     - Formato de colores y grosores de las aristas según el peso, facilitando la interpretación de las conexiones más importantes.

3. **Resultado final**:
   - La red final fue visualmente comparada con la red objetivo para verificar su similitud.
   - La representación gráfica incluye todos los elementos clave de la red, replicando de manera precisa los detalles del grafo proporcionado como referencia.

### Propósito de la visualización
La visualización de la red es crucial para identificar patrones de conectividad, relaciones importantes y características estructurales de la red de manera más intuitiva que con la matriz de adyacencia. Esto permite validar que la red generada cumple con las especificaciones requeridas.




```python
# Visualización Inicial

plt.figure(figsize=(8, 6))

# Posición de los nodos
pos = nx.spring_layout(G)

# Tamaño de los nodos basado en grado (opcional)
node_sizes = [700 + 300 * G.degree(n) for n in G.nodes]

# Grosor de las aristas basado en los pesos
edge_widths = [2 + 5 * d['weight'] for _, _, d in G.edges(data=True)]

# Dibujar nodos
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='skyblue', alpha=0.8)

# Dibujar etiquetas de los nodos
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

# Dibujar aristas
nx.draw_networkx_edges(G, pos, width=edge_widths, edge_color='gray', alpha=0.6)

# Extraer y dibujar etiquetas de los pesos de las aristas
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)

# Título y configuración adicional
plt.title("Visualización Inicial")
plt.axis('off')  # Ocultar ejes
plt.show()
```

### Interpretación de la visualización inicial

La visualización inicial de la red permite analizar la estructura y las relaciones de los nodos. A continuación, se detalla la interpretación:

1. **Nodos**:
   - Los nodos están representados como círculos azules, cuyo tamaño es proporcional al número de conexiones (grado) que poseen.
   - Esto permite identificar rápidamente los nodos más importantes en términos de conectividad dentro de la red.

2. **Aristas (enlaces)**:
   - Las líneas grises entre los nodos representan las conexiones o enlaces de la red.
   - El grosor de las aristas es proporcional al peso de la conexión, lo que ayuda a destacar las relaciones más fuertes. Por ejemplo:
     - El enlace entre los nodos **2 y 4**, con un peso de **0.7**, es la conexión más fuerte y está representado con una arista más gruesa.
     - Los enlaces con pesos más bajos, como **0.1** entre los nodos **2 y 3**, están representados con aristas más delgadas.

3. **Etiquetas**:
   - Cada nodo está etiquetado con su número correspondiente para identificarlo fácilmente.
   - Los pesos de las aristas están anotados cerca de las conexiones, indicando la intensidad de cada relación.

4. **Distribución espacial**:
   - La disposición de los nodos se basa en un algoritmo de diseño (`spring_layout`) que intenta espaciar los nodos uniformemente, minimizando la superposición de enlaces y mejorando la legibilidad.


### Visualización de la red objetivo


Esta representación final de la red permite validar que la disposición de los nodos, las conexiones y los pesos replican fielmente la red objetivo. Además, el uso de aristas curvadas y etiquetas personalizadas asegura una representación visual clara y profesional.

### Observaciones

- Las conexiones fuertes (como **2 y 4**, peso **0.7**) destacan por su prominencia.
- La configuración manual de los nodos y las aristas curvadas ofrece flexibilidad para ajustarse a redes complejas.

Esta visualización cumple con el objetivo de replicar y analizar la red objetivo proporcionada, sirviendo como base para evaluaciones adicionales.


```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear el grafo y asignar los pesos corregidos
G = nx.Graph()
G.add_weighted_edges_from([
    (1, 2, 0.5), (1, 3, 0.2), (1, 5, 0.3),  # Ajuste de peso en 1-5 (curva)
    (2, 3, 0.4), (2, 4, 0.7), (2, 5, 0.4),
    (3, 4, 0.2), (3, 5, 0.2),
    (4, 5, 0.1)
])

# Fijar posiciones manuales de los nodos
pos = {
    1: (0, 1),   # Nodo 1 (parte superior)
    2: (-0.04, 0.5),  # Nodo 2 (izquierda)
    3: (0.03, 0.7),   # Nodo 3 (derecha)
    4: (0.015, 0.35),  # Nodo 4 (abajo a la derecha)
    5: (0, 0)  # Nodo 5 (abajo a la izquierda)
}

# Lista de aristas sin la conexión 1-5 (vertical)
edges_to_draw = [(u, v) for u, v in G.edges if (u, v) != (1, 5) and (v, u) != (1, 5)]

# Dibujar nodos y etiquetas
plt.figure(figsize=(8, 7))
nx.draw_networkx_nodes(G, pos, node_color='white', edgecolors='black', node_size=1000)
nx.draw_networkx_labels(G, pos, font_size=15, font_color='blue', font_weight='bold')

# Dibujar aristas rectas con grosor dinámico
edge_widths = [2 + 5 * G[u][v]['weight'] for u, v in edges_to_draw]
nx.draw_networkx_edges(G, pos, edgelist=edges_to_draw, width=edge_widths, edge_color='black')

# Dibujar aristas curvadas manualmente
arc_edges = [(3, 5), (1, 5)]  # Aristas que queremos curvar
curvatures = [-0.4, -1.5]  # Radios de curvatura
for (u, v), rad in zip(arc_edges, curvatures):
    x1, y1 = pos[u]
    x2, y2 = pos[v]
    weight = G[u][v]['weight']  # Obtener peso de la arista
    # Dibujar la arista curva con grosor dinámico
    plt.annotate(
        '', xy=(x2, y2), xytext=(x1, y1),
        arrowprops=dict(
            arrowstyle='-', color='black', shrinkA=15, shrinkB=15,
            lw=2 + 5 * weight,  # Ajustar el grosor dinámico según el peso
            connectionstyle=f"arc3,rad={rad}"
        )
    )
    # Ajustar etiqueta de peso para arista curva
    if (u, v) == (3, 5):  # Ajuste para la conexión 5 y 3
        offset_x, offset_y = 0.01, 0.03
    elif (u, v) == (1, 5):  # Ajuste para la conexión curva 1 y 5
        offset_x, offset_y = 0.04, -0.04
    else:
        offset_x, offset_y = (0.05 if rad < 0 else -0.05, -0.02 if rad < 0 else 0.02)

    mid_x, mid_y = (x2 + x1) / 2, (y1 + y2) / 2  # Punto medio
    plt.text(
        mid_x + offset_x, mid_y + offset_y, f'{weight:.1f}',
        color='red', fontsize=12, ha='center'
    )

# Dibujar etiquetas de las aristas rectas
labels = {k: v for k, v in nx.get_edge_attributes(G, 'weight').items() if k not in arc_edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=12, font_color='red')

plt.title("Visualización red Objetivo")
plt.axis("off")
plt.show()

```
